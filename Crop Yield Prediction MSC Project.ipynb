```python
#Summary Statistics
import pandas as pd
Main_Data = pd.read_csv('MAINDATA.csv')
Main_Data.head()
```

```python
Main_Data.describe()
```

```python
clean_data = Main_Data.dropna().drop(['Year', 'Region'], axis=1)
clean_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
# get data 

X = clean_data[['Maize', 'Rice', 'Cassava', 'Yam', 'Plantain', 'Tomatoes', \
               'Temperature', 'Pesticides', 'Rainfall', 'N2O']].values#.values
Y = clean_data['Yields'].values.astype('int')

# Splitting the dataset into train and test
X_train, X_test, y_train, y_test = train_test_split( 
X, Y, test_size = 0.3, random_state = 100)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Function to calculate accuracy
def cal_accuracy(y_test, y_pred):
      
    acc = accuracy_score(y_test,y_pred)*100
    return acc
      
```

```python

# Function to make predictions
def prediction(X_test, clf_object):
  
    # Predicton on test with GINI Index
    y_pred = clf_object.predict(X_test)
    print("Predicted values:")
    print(y_pred)
    return y_pred
```

```python
from sklearn import metrics

def metrics_calc(yt, yp):
    # Performance metrics for Crop Decision Tree
    print('Mean Absolute Error:', metrics.mean_absolute_error(yt, yp))
    print('Mean Squared Error:', metrics.mean_squared_error(yt, yp))
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(yt, yp)))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error
```

```python
      
# Function to perform training with giniIndex.
def train_using_gini(X_train, X_test, y_train):
  
    # Creating the classifier object
    clf_gini = DecisionTreeClassifier(criterion = "gini",
            random_state = 100,max_depth=3, min_samples_leaf=5)
  
    # Performing training
    clf_gini.fit(X_train, y_train)
    return clf_gini
      
```

```python

clf_gini = train_using_gini(X_train, X_test, y_train)

# Operational Phase
print("Results Using GINI Index:")
# Prediction using gini
y_pred_gini = prediction(X_test, clf_gini)
print('\naccuracy for gini ', cal_accuracy(y_test, y_pred_gini))

print()
# gini_rmse = mean_squared_error(y_test, y_pred_gini, squared=False)
# print('rmse for GINI index ', gini_rmse)
print('--rmse for GINI index--')
metrics_calc(y_test, y_pred_gini)
```

```python
import numpy as np
```

```python

# Function to perform training with entropy.
def train_using_entropy(X_train, X_test, y_train):
  
    # Decision tree with entropy
    clf_entropy = DecisionTreeClassifier(
            criterion = "entropy", random_state = 100,
            max_depth = 3, min_samples_leaf = 5)
  
    # Performing training
    clf_entropy.fit(X_train, y_train)
    return clf_entropy
```

```python
clf_entropy = train_using_entropy(X_train, X_test, y_train)
print("Results Using Entropy:")
# Prediction using entropy
y_pred_entropy = prediction(X_test, clf_entropy)
print('\nAccuracy for entropy ', cal_accuracy(y_test, y_pred_entropy))

print()
# entropy_rmse = np.sqrt(mean_squared_error(y_test, y_pred_entropy, squared=False))
# print('rmse for decision tree entropy ', entropy_rmse)

print('-- metric for decision tree entropy --')
metrics_calc(y_test, y_pred_entropy)
```

# KNEIGHBORS

```python

#K-Nearest Neighbor
# Import necessary modules
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt

neighbors = np.arange(1, 9)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
  
# Loop over K values
for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
      
    # Compute training and test data accuracy
    train_accuracy[i] = knn.score(X_train, y_train)
    test_accuracy[i] = knn.score(X_test, y_test)
  
    # Generate plot
    
plt.plot(neighbors, test_accuracy, label = 'Testing dataset Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training dataset Accuracy')
  
plt.legend()
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.show()

# print('\naccuracy for KNeighbors ', cal_accuracy(y_test, kneigh_y_pred))

print('train_accuracy', np.max(train_accuracy), 'at index ',np.argmax(train_accuracy) )
print('test_accuracy',np.max(test_accuracy),'at index ',np.argmax(test_accuracy) )
```

```python

#K-Nearest Neighbor

kneigh_clf = KNeighborsClassifier(n_neighbors=2)
kneigh_clf.fit(X_train, y_train)

kneigh_y_pred = kneigh_clf.predict(X_test)

print('\naccuracy for KNeighbors ', cal_accuracy(y_test, kneigh_y_pred))
# print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, kneigh_y_pred)))
print('--metrics for KNeighbors --')
metrics_calc(y_test, kneigh_y_pred)

```
